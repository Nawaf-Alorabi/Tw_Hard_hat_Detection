```python
!pip install ultralytics

from ultralytics import YOLO

model = YOLO("yolov8n-obb.pt")

model.train(
    data="/content/Hat-1/data.yaml",
    epochs=20,
    imgsz=640
)
```

```python
from ultralytics import YOLO
from IPython.display import display, Image
import os

# =========================
# 1) Evaluation (الأرقام)
# =========================
metrics = model.val()

print("=== Validation Metrics ===")
print(f"Precision: {metrics.box.mp:.3f}")
print(f"Recall: {metrics.box.mr:.3f}")
print(f"mAP50: {metrics.box.map50:.3f}")
print(f"mAP50-95: {metrics.box.map:.3f}")

# =========================
# 2) Test Prediction
# =========================
results = model.predict(
    source="/content/Hat-1/test/images",
    save=True,
    conf=0.25,
    verbose=False
)

pred_dir = results[0].save_dir

# =========================
# 3) عرض ملخص لكل صورة
# =========================
print("\n=== Test Predictions Summary ===\n")

for i, r in enumerate(results, 1):
    class_ids = r.obb.cls.cpu().numpy().astype(int) if r.obb is not None else []
    names = [r.names[c] for c in class_ids]

    counts = {}
    for name in names:
        counts[name] = counts.get(name, 0) + 1

    print(f"Image {i}: {os.path.basename(r.path)}")
    if counts:
        print(counts)
    else:
        print("No detections")
    print()

# =========================
# 4) عرض بعض الصور (5 فقط)
# =========================
print("\n=== Sample Predictions ===\n")

for r in results[:5]:
    img_path = os.path.join(pred_dir, os.path.basename(r.path))
    print(os.path.basename(r.path))
    display(Image(filename=img_path, width=400))
```

```python
!pip install ultralytics

from ultralytics import YOLO

model = YOLO("yolov8s-obb.pt")

model.train(
    data="/content/Hat-1/data.yaml",
    epochs=25,
    imgsz=640,
    batch=8,
    patience=20,
    optimizer="AdamW",
    lr0=0.001,
    fliplr=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10,
    translate=0.1,
    scale=0.2
)
```

```python
from ultralytics import YOLO
from IPython.display import display, Image
import os

# =========================
# 1) Evaluation
# =========================
metrics = model.val()

print("=== Tuned Model Validation Metrics ===")
print(f"Precision: {metrics.box.mp:.3f}")
print(f"Recall: {metrics.box.mr:.3f}")
print(f"mAP50: {metrics.box.map50:.3f}")
print(f"mAP50-95: {metrics.box.map:.3f}")

# =========================
# 2) Test Prediction
# =========================
results = model.predict(
    source="/content/Hat-1/test/images",
    save=True,
    conf=0.25,
    verbose=False
)

pred_dir = results[0].save_dir

# =========================
# 3) Summary per image
# =========================
print("\n=== Tuned Model Test Predictions Summary ===\n")

for i, r in enumerate(results, 1):
    class_ids = r.obb.cls.cpu().numpy().astype(int) if r.obb is not None else []
    names = [r.names[c] for c in class_ids]

    counts = {}
    for name in names:
        counts[name] = counts.get(name, 0) + 1

    print(f"Image {i}: {os.path.basename(r.path)}")
    if counts:
        print(counts)
    else:
        print("No detections")
    print()

# =========================
# 4) Show sample predictions
# =========================
print("\n=== Tuned Model Sample Predictions ===\n")

for r in results[:5]:
    img_path = os.path.join(pred_dir, os.path.basename(r.path))
    print(os.path.basename(r.path))
    display(Image(filename=img_path, width=400))
```
